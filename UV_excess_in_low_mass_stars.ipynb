## Introduction

**Light curve background**: a light curve represents the brightness variation of an object studied over a period of time. This tool is simple but powerful, especially useful for time-variable objects such as supernovae or variable stars. The plot can show brightness (magnitude) over time, but it can also be represented as any other measure of brightness, such as the count of the number of X-rays over time.

**Defining some terms**:
- **csv file**: comma-separated values file. Text file in which each row is separated by commas, it is widely used in the astro-community and allows data to be saved with the structure of a table.

#### Download the light curve using _gPhoton_

_gAperture_ is the photometry tool of _gPhoton_. It performes aperture photometry at the given sky position and returns the light curves at a specified bin width.

```python
gPhoton.gAperture(band='NUV', skypos=[176.91975, 0.25561], stepsz=10.,
          csvfile='gj_3685a_lc.csv', radius=0.012,
          annulus=[0.013,0.016],
          trange=[766525335.,766526573.])
```

#### Reading the _csv_ file using _pandas_

The _pandas_ package is very useful for reading _csv_ files since we can directly use the _read_csv_, more information about this command can be found <a href="https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html">here</a>.

```python
data= pd.read_csv("gj_3685a_lc.csv")
data
```

#### Plotting the light curve

Once we have loaded the _csv_ file, we can create a figure of the flux versus the time in the following way:

```python
plt.figure()
plt.plot(data.t0 - data.t0[0],data.flux_bgsub/1e-13,'ko')
plt.plot(data.t0 - data.t0[0],data.flux_bgsub/1e-13,'k')
plt.xlabel('GALEX Time - %.3f'%data.t0[0])
plt.ylabel('Flux ($ergs/sec/cm^2/\AA$/1E-13)')
plt.show()
```

**Note**: we have chosen to substract the initial time from the time vector based on convention and for clarity purposes, but this may be altered based on the user's choice.
